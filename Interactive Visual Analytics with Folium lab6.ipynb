```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# 1. Define the Proximities (Lat, Lon)
proximities = {
    'Railway': [28.57211, -80.58521],
    'Highway': [28.56321, -80.57079],
    'City': [28.61220, -80.80750]
}

launch_site = [28.56230, -80.57736]

# 2. Loop through proximities to calculate distance and draw lines
for name, prox_coord in proximities.items():
    # Calculate distance
    dist = calculate_distance(launch_site[0], launch_site[1], prox_coord[0], prox_coord[1])
    
    # Add Distance Marker
    folium.Marker(
        location=prox_coord,
        icon=DivIcon(
            icon_size=(150,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#2E8B57;"><b>%s: %s</b></div>' % (name, "{:10.2f} KM".format(dist)),
        )
    ).add_to(site_map)
    
    # Draw Line
    folium.PolyLine(
        locations=[launch_site, prox_coord], 
        weight=1, 
        color='blue'
    ).add_to(site_map)

# Show the map
site_map
```
